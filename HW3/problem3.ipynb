```python
from collections import deque

# Start the graph search from root using a doubly-ended queue

def DFS(graph, root):

    # deque is list-like container with fast appends and pops on either end
    visited, queue = list(), deque([root])
    visited.append(root)

    print('You took this path.')

    # Keep exploring as long as the queue is not empty
    while queue:
        # Dequeue a vertex from queue
        vertex = queue.pop()
        print(vertex, end=" ")

        # when searching reaches 49, stop the process and print out successful message
        if vertex == 49:
            print('\nCONGRATULATION: You SUCCESSFULLY found the exit')
            break

        # If not visited, mark it as visited, and enqueue it
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)

    # if the last visited point is not 49, print out unsuccessful message
    if visited.pop() != 49:
        print('\nSORRY! You are UNSUCCESSFUL! This is not a path to exit.')
```
