```python
from collections import deque

# Start the graph search from root using a doubly-ended queue

def FindFriends(graph, root, level):

    # deque is list-like container with fast appends and pops on either end
    queue = deque([root])

    # use dict to set depth of root == 0
    depths = {root : 0}

    # Keep exploring as long as the queue is not empty
    while queue:
        # Dequeue a vertex from queue
        vertex = queue.popleft()

        # set the same new depth for all neighbors, and enqueue them
        for neighbor in graph[vertex]:
            if neighbor in depths:
              continue
            queue.append(neighbor)
            depths[neighbor] = depths[vertex] + 1

    # Print out all friends at a certain level traced from root
    print(f'Friends of {root} at level {level}:')
    for x, y in depths.items():
      if y == level:
        print(x)



```
